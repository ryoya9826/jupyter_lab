```python
def kazarimozi(s, sequence):# s は wordの空白区切りを想定，各wordの文頭に（TeX記法で）\sequence をつけたものを返してほしい
    l = s.split()
    l1=[]
    for x in l:
        x= '\\' + sequence + '{' + x[0] + '}' +x[1:]
        l1.append(x)
    return ' '.join(l1)
    
print(kazarimozi('Cohen Macaulay', 'textbf'))
assert kazarimozi('Cohen Macaulay', 'textbf') == '\\textbf{C}ohen \\textbf{M}acaulay'

# '\\textbf{C}ohen \\textbf{M}acaulay'を返してくる.
```
