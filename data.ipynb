```python
def str_reverse(s):
    return ''.join(reversed(s))

def my_round(n, digit=0):
    p = 10**(-digit-1)
    return int((n*p*2 + 1) // 2 / p)

def SumOfDigits(n):
    s=0
    while n>0:
        s=s+ n%10
        n= n//10
    return s

from decimal import *

def DecimalRound(n,d):
    digit = '1E' + str(d)
    if d >= 0:
        return int(Decimal(n).quantize(Decimal(digit), rounding=ROUND_HALF_UP))
    else:
        return Decimal(n).quantize(Decimal(digit), rounding=ROUND_HALF_UP)

def de_floor(n): #小数点以下切り捨て
    return int(Decimal(n).quantize(Decimal('0'), rounding=ROUND_FLOOR))


def TransposeRotate(l): #転置
    h, w = len(l), len(l[0])
    ans = [[None] * h for _ in range(w)]
    for i in range(w):
        for j in range(h):
            ans[i][j] = l[j][i]

    return ans

def succ(str,num=1):#文字の後続関数
    str=str.lower()
    l=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
    return l[(l.index(str)+num)%26]

def succ(num):#数字の後続関数
    if num == 9:
        return 0
    else:
        return num + 1

def Replace(S,i):
    return S[:i] + S[i+1] + S[i] + S[i+2:]

def int_to_str(n): # 1 <= n <= 26
    return chr(n-1+97)
```

```python
#2点間の距離と角度
def dist(p1,p2):
    return math.sqrt((p1[0]-p2[0]) ** 2 + (p1[1]-p2[1]) ** 2)

# v1,v2はベクトル
# u = [s[0]-x[0],s[1]-x[1]]
# v = [t[0]-x[0],t[1]-x[1]]
# とかしてangle(u,v)でやる

def angle(v1,v2):
    i = v1[0]*v2[0]+v1[1]*v2[1]
    n = dist([0,0],v1) * dist([0,0],v2)
    c = i/n
    if c > 1:
        c = 1
    if c < -1:
        c = -1
    return math.acos(c)*180/math.pi
```
