```python
#https://atcoder.jp/contests/abc275/tasks/abc275_c
import math

def dist(p1,p2):
    return math.sqrt((p1[0]-p2[0]) ** 2 + (p1[1]-p2[1]) ** 2)

def angle(p1,p2):
    i = p1[0]*p2[0]+p1[1]*p2[1]
    n = dist([0,0],p1) * dist([0,0],p2)
    c = i/n
    if c > 1:
        c = 1
    if c < -1:
        c = -1
    return math.acos(c)*180/math.pi

S=[input() for _ in range(9)]
l=[]
for i in range(9):
    for j in range(9):
        if S[i][j] == '#':
            l.append([i,j])
ans_D=[]
for x in l:
    for s in l:
        for t in l:
            if s != x and t != x:
                u = [s[0]-x[0],s[1]-x[1]]
                v = [t[0]-x[0],t[1]-x[1]]
                if dist(x,s) == dist(x,t) and angle(u,v) == 90:
                    for y in l:
                        u = [s[0]-y[0],s[1]-y[1]]
                        v = [t[0]-y[0],t[1]-y[1]]
                        if y != x and dist(y,s) == dist(x,s) and angle(u,v) == 90:
                            ans_D.append([x,s,t,y])

for i in range(len(ans_D)):
    ans_D[i].sort()
    ans_D[i]=[tuple(x) for x in ans_D[i]]
A = set(map(tuple,ans_D))
print(len(A))
```

```python
#https://atcoder.jp/contests/abc276/tasks/abc276_c
import math
N=int(input())
P=list(map(int,input().split()))

#順列Pが何番目かを計算する

cnt = 0
chk = []
for i in range(len(P)):
    tmp_c = 0
    for j in range(1,P[i]):
        if j not in chk:
            tmp_c += 1
    chk.append(P[i])
    cnt += tmp_c*math.factorial((len(P)-1-i))

#cnt-1番目の順列を求める
#参考:https://itakanaya9.hatenablog.com/entry/2014/02/17/121428

l=[]
cnt -= 1
for i in range(1,N+1):
    q,r=divmod(cnt,i)
    cnt = q
    l.append(r)
l.reverse()
ans=[]
S=[i for i in range(1,N+1)] 
for x in l:
    ans.append(S[x])
    del S[x]
print(*ans)
```

```python
#https://atcoder.jp/contests/abc270/tasks/abc270_c
from collections import deque
N,X,Y=map(int,input().split())
tree=[[] for _ in range(N+1)]
for _ in range(N-1):
    x,y=map(int,input().split())
    tree[x].append(y)
    tree[y].append(x)
upper=[0]*(N+1)
upper[X] = -1
q = deque()
q.append(X)
while q:
    now =q.popleft()
    for nxt in tree[now]:
        if upper[nxt] == 0:
            upper[nxt] = now
            q.append(nxt)
ans = []
ans.append(Y)
while True:
    if Y == X or Y == -1:
        break
    else:
        ans.append(upper[Y])
        Y=upper[Y]
ans=list(reversed(ans))
print(*ans)
```

```python
#https://atcoder.jp/contests/abc264/tasks/abc264_c
import itertools
    
def TransposeRotate(M): #転置
    H,W = len(M),len(M[0])
    ans = [[None]*H for _ in range(W)]
    for i in range(W):
        for j in range(H):
            ans[i][j] = M[j][i]
    return ans

H,W = map(int,input().split())
M = [list(map(int,input().split())) for _ in range(H)]
h,w = map(int,input().split())
B = [list(map(int,input().split())) for _ in range(h)]

A = list(map(list,list(itertools.combinations(M,h))))
A = list(map(TransposeRotate,A))
tmp = []
for x in A:
    tmp += list(map(list,list(itertools.combinations(x,w))))
A = list(map(TransposeRotate,tmp))

if B in A:
    print('Yes')
else:
    print('No')
```

```python
#https://atcoder.jp/contests/abc259/tasks/abc259_c
def split_connect(S):
    before = ''
    tmp = S[0]
    l = []
    for x in list(S):
        if before != x and before != '':
            l.append(tmp)
            tmp = ''
        tmp += x
        before = x
    else:
        l.append(tmp)
    return l
            
S=input()
T=input()
Sc = split_connect(S)
Tc = split_connect(T)
if len(S) >= len(T) and S != T:
    print('No')
    exit()
if len(Sc) == len(Tc):
    for i in range(len(Sc)):
        if (Sc[i][0] != Tc[i][0]) or (len(Sc[i]) == 1 and len(Tc[i]) > 1) or (len(Sc[i]) > len(Tc[i])):
            print('No')
            break
    else:
        print('Yes')
else:
    print('No')
    
#############################################################
#上は素朴な解法．
#RLE (Run Length Encoding) を使った実装を下に書いてみる
#（しかしRLEのほうが低速（実装が悪い？）
def RLE(S):
    before = ''
    cnt = 0
    tmp = S[0]
    l = []
    for x in list(S):
        if before != x and before != '':
            l.append([tmp,cnt])
            tmp = x
            cnt = 0
        cnt += 1
        before = x
    else:
        l.append([x,cnt])
    return l
            
S=input()
T=input()
Rs = RLE(S)
Rt = RLE(T)

if len(Rs) == len(Rt):
    for i in range(len(Rs)):
        if Rs[i][0] != Rt[i][0] or (Rs[i][1] == 1 and Rt[i][1] > 1) or (Rs[i][1] > Rt[i][1]):
            print('No')
            break
    else:
        print('Yes')
else:
    print('No')
```

```python
#https://atcoder.jp/contests/abc257/tasks/abc257_c
#人数が同じ場合の処理は参考:https://blog.hamayanhamayan.com/entry/2022/06/25/225842
import itertools
N = int(input())
S = list(map(int,list(input())))
W = list(map(int,input().split()))
l = []
for i in range(N):
    l.append([W[i],S[i]])
l = sorted(l, reverse=True, key=lambda x: x[1])
l = sorted(l, key=lambda x: x[0])
L = []
for i in range(N):
    L.append(l[i][1])
Wrong_A = [0]
Wrong_A += list(itertools.accumulate(L))
Cr = list(itertools.accumulate(reversed(L)))
Wrong = []
for i in range(N+1):
    if i != N:
        Wrong.append(Wrong_A[i]+N-i-Cr[N-i-1])
    else:
        Wrong.append(Wrong_A[i])
print(N-min(Wrong))
```
