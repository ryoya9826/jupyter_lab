```python
#https://atcoder.jp/contests/abc275/tasks/abc275_c
import math

def dist(p1,p2):
    return math.sqrt((p1[0]-p2[0]) ** 2 + (p1[1]-p2[1]) ** 2)

def angle(p1,p2):
    i = p1[0]*p2[0]+p1[1]*p2[1]
    n = dist([0,0],p1) * dist([0,0],p2)
    c = i/n
    if c > 1:
        c = 1
    if c < -1:
        c = -1
    return math.acos(c)*180/math.pi

S=[input() for _ in range(9)]
l=[]
for i in range(9):
    for j in range(9):
        if S[i][j] == '#':
            l.append([i,j])
ans_D=[]
for x in l:
    for s in l:
        for t in l:
            if s != x and t != x:
                u = [s[0]-x[0],s[1]-x[1]]
                v = [t[0]-x[0],t[1]-x[1]]
                if dist(x,s) == dist(x,t) and angle(u,v) == 90:
                    for y in l:
                        u = [s[0]-y[0],s[1]-y[1]]
                        v = [t[0]-y[0],t[1]-y[1]]
                        if y != x and dist(y,s) == dist(x,s) and angle(u,v) == 90:
                            ans_D.append([x,s,t,y])

for i in range(len(ans_D)):
    ans_D[i].sort()
    ans_D[i]=[tuple(x) for x in ans_D[i]]
A = set(map(tuple,ans_D))
print(len(A))
```

```python
#https://atcoder.jp/contests/abc259/tasks/abc259_c
def split_connect(S):
    before = ''
    tmp = S[0]
    l = []
    for x in list(S):
        if before != x and before != '':
            l.append(tmp)
            tmp = ''
        tmp += x
        before = x
    else:
        l.append(tmp)
    return l
            
S=input()
T=input()
Sc = split_connect(S)
Tc = split_connect(T)
if len(S) >= len(T) and S != T:
    print('No')
    exit()
if len(Sc) == len(Tc):
    for i in range(len(Sc)):
        if (Sc[i][0] != Tc[i][0]) or (len(Sc[i]) == 1 and len(Tc[i]) > 1) or (len(Sc[i]) > len(Tc[i])):
            print('No')
            break
    else:
        print('Yes')
else:
    print('No')
    
#############################################################
#上は素朴な解法．
#RLE (Run Length Encoding) を使った実装を下に書いてみる
#（しかしRLEのほうが低速（実装が悪い？）
def RLE(S):
    before = ''
    cnt = 0
    tmp = S[0]
    l = []
    for x in list(S):
        if before != x and before != '':
            l.append([tmp,cnt])
            tmp = x
            cnt = 0
        cnt += 1
        before = x
    else:
        l.append([x,cnt])
    return l
            
S=input()
T=input()
Rs = RLE(S)
Rt = RLE(T)

if len(Rs) == len(Rt):
    for i in range(len(Rs)):
        if Rs[i][0] != Rt[i][0] or (Rs[i][1] == 1 and Rt[i][1] > 1) or (Rs[i][1] > Rt[i][1]):
            print('No')
            break
    else:
        print('Yes')
else:
    print('No')
```

```python
#https://atcoder.jp/contests/abc253/tasks/abc253_c
import math
from bisect import bisect_left, bisect_right
from typing import Generic, Iterable, Iterator, TypeVar, Union, List
T = TypeVar('T')

class SortedSet(Generic[T]):
    BUCKET_RATIO = 50
    REBUILD_RATIO = 170

    def _build(self, a=None) -> None:
        "Evenly divide `a` into buckets."
        if a is None: a = list(self)
        size = self.size = len(a)
        bucket_size = int(math.ceil(math.sqrt(size / self.BUCKET_RATIO)))
        self.a = [a[size * i // bucket_size : size * (i + 1) // bucket_size] for i in range(bucket_size)]
    
    def __init__(self, a: Iterable[T] = []) -> None:
        "Make a new SortedSet from iterable. / O(N) if sorted and unique / O(N log N)"
        a = list(a)
        if not all(a[i] < a[i + 1] for i in range(len(a) - 1)):
            a = sorted(set(a))
        self._build(a)

    def __iter__(self) -> Iterator[T]:
        for i in self.a:
            for j in i: yield j

    def __reversed__(self) -> Iterator[T]:
        for i in reversed(self.a):
            for j in reversed(i): yield j
    
    def __len__(self) -> int:
        return self.size
    
    def __repr__(self) -> str:
        return "SortedSet" + str(self.a)
    
    def __str__(self) -> str:
        s = str(list(self))
        return "{" + s[1 : len(s) - 1] + "}"

    def _find_bucket(self, x: T) -> List[T]:
        "Find the bucket which should contain x. self must not be empty."
        for a in self.a:
            if x <= a[-1]: return a
        return a

    def __contains__(self, x: T) -> bool:
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        return i != len(a) and a[i] == x

    def add(self, x: T) -> bool:
        "Add an element and return True if added. / O(√N)"
        if self.size == 0:
            self.a = [[x]]
            self.size = 1
            return True
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i != len(a) and a[i] == x: return False
        a.insert(i, x)
        self.size += 1
        if len(a) > len(self.a) * self.REBUILD_RATIO:
            self._build()
        return True

    def discard(self, x: T) -> bool:
        "Remove an element and return True if removed. / O(√N)"
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i == len(a) or a[i] != x: return False
        a.pop(i)
        self.size -= 1
        if len(a) == 0: self._build()
        return True
    
    def lt(self, x: T) -> Union[T, None]:
        "Find the largest element < x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] < x:
                return a[bisect_left(a, x) - 1]

    def le(self, x: T) -> Union[T, None]:
        "Find the largest element <= x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] <= x:
                return a[bisect_right(a, x) - 1]

    def gt(self, x: T) -> Union[T, None]:
        "Find the smallest element > x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] > x:
                return a[bisect_right(a, x)]

    def ge(self, x: T) -> Union[T, None]:
        "Find the smallest element >= x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] >= x:
                return a[bisect_left(a, x)]
    
    def __getitem__(self, x: int) -> T:
        "Return the x-th element, or IndexError if it doesn't exist."
        if x < 0: x += self.size
        if x < 0: raise IndexError
        for a in self.a:
            if x < len(a): return a[x]
            x -= len(a)
        raise IndexError
    
    def index(self, x: T) -> int:
        "Count the number of elements < x."
        ans = 0
        for a in self.a:
            if a[-1] >= x:
                return ans + bisect_left(a, x)
            ans += len(a)
        return ans

    def index_right(self, x: T) -> int:
        "Count the number of elements <= x."
        ans = 0
        for a in self.a:
            if a[-1] > x:
                return ans + bisect_right(a, x)
            ans += len(a)
        return ans

from collections import defaultdict
S=defaultdict(int)
Q=int(input())
s=SortedSet()
for _ in range(Q):
    q=list(map(int,input().split()))
    if q[0] == 1:
        S[q[1]] += 1
        s.add(q[1])
    elif q[0] == 2:
        S[q[1]] -= min(q[2],S[q[1]])
        if S[q[1]] <= 0:
            s.discard(q[1])
    else:
        print(s[len(s)-1]-s[0])
```

```python
#https://atcoder.jp/contests/abc021/tasks/abc021_c
from collections import defaultdict, deque
graph = defaultdict(list)
depth = defaultdict(int)
N = int(input())
a,b = map(int,input().split())
M = int(input())
for _ in range(M):
    x,y = map(int,input().split())
    graph[x].append(y)
    graph[y].append(x)
q = deque()
q.append(a)
visited = {a}
depth[a] = 0
while q:
    now = q.popleft()
    for nxt in graph[now]:
        if nxt not in visited:
            visited.add(nxt)
            depth[nxt] = depth[now] + 1
            q.append(nxt)
d =depth[b]
n_graph=defaultdict(list)
before = [b]
while before != []:
    for x in before:
        n = depth[x]
        tmp = []
        for y in graph[x]:
            if depth[y] == n-1:
                n_graph[y].append(x)
                n_graph[x].append(y)
                tmp.append(y)
        before = tmp
graph = n_graph
width = defaultdict(int)
q = deque()
q.append(a)
visited = {a}
width[0] += 1
while q:
    now = q.popleft()
    for nxt in graph[now]:
        if nxt not in visited:
            visited.add(nxt)
            width[depth[nxt]] += 1
            q.append(nxt)
ans = 1
for x in width:
    ans *= width[x]
ans //= width[d]
print(ans%1000000007)
```

```python
#https://atcoder.jp/contests/abc278/tasks/abc278_f
N = int(input())
S = [input() for _ in range(N)]
dp = [[0] * (2**N) for _  in range(N)]

def solve(last,state)-> None:
    "1,-1のどれかを返す．1なら黒勝ち"
    if dp[last][state] != 0:
        return dp[last][state]
    last_char = S[last][-1]
    if bin(state).count('1')%2 != 0:
        # 白番
        for i in range(N):
            if state & (1 << i) or last_char != S[i][0]:
                continue
            if solve(i,state | (1 << i)) == -1:
                dp[last][state] = -1
                return -1
        dp[last][state] = 1
        return 1
    else:
        # 黒番
        for i in range(N):
            if state & (1 << i) or last_char != S[i][0]:
                continue
            if solve(i,state | (1 << i)) == 1:
                dp[last][state] = 1
                return 1
        dp[last][state] = -1
        return -1

for i in range(N):
    if solve(i,1<<i) == 1:
        print('First')
        break
else:
    print("Second")
```

```python
#https://atcoder.jp/contests/dp/tasks/dp_f
S = input()
T = input()
N = len(S)
M = len(T)
dp = [[0 for _ in range(M+1)] for _  in range(N+1)]
for i in range(N):
    for j in range(M):
        if S[i] == T[j]:
            dp[i+1][j+1] = max(dp[i][j]+1,dp[i+1][j],dp[i][j+1])
        else:
            dp[i+1][j+1] = max(dp[i+1][j],dp[i][j+1])
ans = []
i = N
j = M
while i > 0 and j > 0:
    if S[i-1] == T[j-1]:
        ans.append(S[i-1])
        i -= 1
        j -= 1
    elif dp[i-1][j] > dp[i][j-1]:
        i -= 1
    else:
        j -= 1

print(''.join(ans[::-1]))
```

```python
#https://atcoder.jp/contests/abc282/tasks/abc282_d
from collections import defaultdict as ddict
from collections import deque

def find(x):
    if parents[x] < 0:
        return x
    else:
        parents[x] = find(parents[x])
        return parents[x]

def union(x, y):
    x = find(x)
    y = find(y)
    if x == y:
        return
    if parents[x] > parents[y]:
        x,y = y,x
    parents[x] += parents[y]
    parents[y] = x

N,M = map(int,input().split())
parents = [-1] *(N)
ans = N*(N-1)//2-M
G = ddict(list)
for _ in range(M):
    u,v = map(int,input().split())
    union(u-1,v-1)
    G[u].append(v)
    G[v].append(u)
root = [i+1 for i, x in enumerate(parents) if x < 0]
connected = ddict(list)
color = [-1]*(N+10) #未探索 -1, black 1, white 0
flag = True
for v in root:
    if color[v] != -1: 
        continue
    q = deque() 
    color[v] = 0
    q.append(v)
    connected[v].append(v)
    while q:
        now = q.popleft()
        for nxt in G[now]:
            if color[nxt] != -1:
                if color[nxt] == color[now]: 
                    flag = False
                continue
            color[nxt] = 1 - color[now]
            q.append(nxt)
            connected[v].append(nxt)
if flag == True:
    for x in root:
        cnt_b = 0
        cnt_w = 0
        for y in connected[x]:
            if color[y] == 1:
                cnt_b += 1
            else:
                cnt_w += 1
        ans -= cnt_b*(cnt_b-1)//2
        ans -= cnt_w*(cnt_w-1)//2
    print(ans)
else:
    print(0)
```

```python
#https://atcoder.jp/contests/abc268/tasks/abc268_d
from itertools import permutations
N,M = map(int,input().split())
S = [input() for _ in range(N)]
T = set([input() for _ in range(M)])
if N == 1:
    if S[0] not in T and len(S[0]) >= 3:
        print(S[0])
    else:
        print(-1)
else:
    L = 16
    for x in S:
        L -= len(x)
    if L >= N-1:
        L += 1 
        Underbar = ['']
        tmp = ''
        for i in range(L):
            tmp += '_'
            Underbar.append(tmp)
        for s in permutations(S,N):
            ans = [['',0]]
            for i in range(N):
                tmp = []
                for x,l in ans:
                    if i < N-1:
                        for k in range(1,L-l):
                            tmp.append([x+s[i]+Underbar[k],l+k])
                    elif x+s[i] not in T:
                        print(x+s[i],len(x+s[i]))
                ans = tmp[:]
        else:
            print(-1)
    else:
        print(-1)                 
```

```python
#https://atcoder.jp/contests/arc037/tasks/arc037_b
#閉路検出の例
class UnionFind():
    def __init__(self, n):
        self.n = n
        self.parents = [-1] * n

    def find(self, x):
        if self.parents[x] < 0:
            return x
        else:
            self.parents[x] = self.find(self.parents[x])
            return self.parents[x]

    def union(self, x, y):
        x = self.find(x)
        y = self.find(y)

        if x == y:
            return

        if self.parents[x] > self.parents[y]:
            x, y = y, x

        self.parents[x] += self.parents[y]
        self.parents[y] = x

    def size(self, x):
        return -self.parents[self.find(x)]

    def same(self, x, y):
        return self.find(x) == self.find(y)

    def members(self, x):
        root = self.find(x)
        return [i for i in range(self.n) if self.find(i) == root]

    def roots(self):
        return [i for i, x in enumerate(self.parents) if x < 0]

    def group_count(self):
        return len(self.roots())

    def all_group_members(self):
        group_members = defaultdict(list)
        for member in range(self.n):
            group_members[self.find(member)].append(member)
        return group_members

N,M = map(int,input().split())
Edges = []
uf = UnionFind(N)
for _ in range(M):
    u,v = map(int,input().split())
    u -= 1
    v -= 1
    uf.union(u,v)
    Edges.append([u,v])
cnt = 0
d = uf.all_group_members()
for mem in d:
    uf2 = UnionFind(N)
    flag = 1
    for u,v in Edges:
        if u in d[mem]:
            if uf2.find(u) == uf2.find(v):
                flag = 0
                break
            else:
                uf2.union(u,v)
    cnt += flag
print(cnt)
```

```python
#答えで二分探索の例
#https://atcoder.jp/contests/typical90/tasks/typical90_a

N,L = map(int,input().split())
K = int(input())
A = list(map(int,input().split()))

def judge(M):
    cnt = 0
    tmp = 0
    for i in range(N):
        if A[i]-tmp >= M and L-A[i] >= M:
            tmp = A[i]
            cnt += 1
    if cnt >= K:
        return True
    else:
        return False
left = -1
right = L+1
while abs(right-left) > 1:
    mid = (left+right)//2
    if judge(mid):
        left = mid
    else:
        right = mid
print(left)
```

```python
#https://atcoder.jp/contests/typical90/tasks/typical90_h
def point(x):
    tmp = list('atcoder')
    if x in 'atcoder':
        return tmp.index(x)+1
    else:
        return -1
mod = 10**9+7
N = int(input())
S = input()
dp = [[0] * (10) for _ in range(N+1)]
dp[0][0] = 1
for i in range(N):
    for j in range(8):
        n = point(S[i])
        if n == j:
            dp[i+1][j] = (dp[i][j] + dp[i][j-1])%mod
        else:
            dp[i+1][j] = dp[i][j]%mod
print(dp[N][7])
```
